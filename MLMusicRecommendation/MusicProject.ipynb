```python
# PARA HACER UN COMENTARIO HAY QUE APRETAR CTRL + /
# PARA HACER RUN LA CELDA SIN QUE VAYA CAMBIANDO HACEMOS CTRL + ENTER
# Va a ser un projecto donde vamos a predecir cual es el genero que le gusta a
# la gente depende de su genero

# El archivo que vamos a usar es uno donde 1 es hombre y 0 es mujer, y estimamos
# los gustos musicales dependiendo de la edad

# Ej: en nuestro data set tenemos que hombre de 20, 23 y 26 le gusta el hiphop,
# pero no tiene dato de 22, asi que nuestro modelo es para que lo prediga.

import pandas as pd

music_data = pd.read_csv('music.csv')
music_data
```

```python
# CLEANING DATA (eliminate duplicates or zeros)
# We are going to separate the input (gender) and output (genre), que tiene nuestra prediccion

X = music_data.drop(columns=['genre']) #aca ya creamos un dataset llamado X sin el genero
y = music_data['genre']
X
```

```python
# SELECT THE ALGORHYTM 
# We are going to use a library

from sklearn.tree import DecisionTreeClassifier       #sklearn(library) tree(module) DecisionTreeClassifier(class) 

# We need to start training this model
model = DecisionTreeClassifier()   #we create the model
model.fit(X.values,y)    #we give it two attributes (I added the ".values"). This is for training
predictions = model.predict([ [21, 1], [22, 0] ])  #takes a two dimensional array, we give it the value of a 21yo male and 22yo female (which we dont have)
predictions
```

```python
# Now we need to calculate the ACCURACY
# Rule of thumb: use 70-80% of data for training, 20-30% for testing.
# Right now in model.fit(X,y) we are passing the entire set, so we need to split the amount
# So instead of passing only two samples (21,1 and 22,0), we can pass the percentage
# we want of the data for testing. 

from sklearn.model_selection import train_test_split #sklearn(library) tree(module) train_test_split(function)

#this will return a tuple
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)  #meaning we are using 20% of data for testing
#Aca divide valores: en 4 tablas diferentes (2 de gender y age, y 2 de solo genre)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

music_data = pd.read_csv('music.csv')
music_data
X = music_data.drop(columns=['genre']) 
y = music_data['genre']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train,y_train)   #le vamos a pasar esa parte nomas, no todo
predictions = model.predict(X_test) #igual aca

#Accuracy is obtained by comparing "predictions(X_test)" with y_test
# so we import:
from sklearn.metrics import accuracy_score

score = accuracy_score(y_test, predictions)#predicted values vs actual values
score     #1.0 means 100%

# Es decir, a la maquina le ensenamos de cuales son los resultados entre ambas tablas de train, entonces queda saber si la tabla
# de input de X_test es exacta con la de ouput de y_test. A medida que vayamos corriendo ira cambiando los valores. Si yo le 
# subo el test_size a 0.8 significa que va a quedar menos data para el train y por ende el accuracy va a bajar mucho.
# A veces va a bajar bastante (aun con test en 0.2) porque en ML para entrenar un modelo se necesita millones de datos. 
```

```python
# Now, we want to do something so we can save our module and don't have to train it every time. 
# Thats why we import a new library that helps to save and upload models. 

import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
import joblib

music_data = pd.read_csv('music.csv')
X = music_data.drop(columns=['genre']) #aca ya creamos un dataset llamado X sin el genero
y = music_data['genre']

model = DecisionTreeClassifier()
model.fit(X.values,y) 

joblib.dump(model, 'music-recommends.joblib')   #we give it two arguments, the name of the model and the name of the file we want to save it 

# predictions = model.predict([ [21, 1], [22, 0] ])
```

```python
# Export our model into a visualization

import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree

music_data = pd.read_csv('music.csv')
X = music_data.drop(columns=['genre']) 
y = music_data['genre']  #has all the classes (types of music, but there are repeated)

model = DecisionTreeClassifier()
model.fit(X.values,y) 

tree.export_graphviz(model, out_file='music-recommends.dot', feature_names=['age', 'gender'], class_names= sorted(y.unique()), label='all', rounded=True, filled=True)
# we run it and then a file was created to be opened with VS Code. 
```
